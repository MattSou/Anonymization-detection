```python
import json
import os
import xml.etree.ElementTree as ET

# Fonction pour créer un élément XML
def create_xml_element(tag, text):
    element = ET.Element(tag)
    element.text = text
    return element

# Charger les annotations du fichier JSON
with open('/home/msouda/Datasets/coco_new_synth_anonymized_bis/annotations/coco_test_annotations.json') as json_file:
    coco_data = json.load(json_file)

# Créer un dictionnaire pour les annotations par image_id
annotations_by_image = {}
for annotation in coco_data['annotations']:
    image_id = annotation['image_id']
    if image_id not in annotations_by_image:
        annotations_by_image[image_id] = []
    annotations_by_image[image_id].append(annotation)

# Créer un dictionnaire pour les images par image_id
images_by_id = {image['id']: image for image in coco_data['images']}

# Créer un dictionnaire pour les catégories par category_id
categories_by_id = {category['id']: category['name'] for category in coco_data['categories']}

# Créer un dossier pour les fichiers XML
os.makedirs('/home/msouda/Datasets/coco_new_synth_anonymized_bis/test/xml_annotations', exist_ok=True)

# Parcourir les annotations par image
for image_id, annotations in annotations_by_image.items():
    image = images_by_id[image_id]
    annotation_filename = f"/home/msouda/Datasets/coco_new_synth_anonymized_bis/test/xml_annotations/{image['file_name'].replace('.jpg', '.xml')}"

    # Créer l'arbre XML
    annotation = ET.Element('annotation')
    ET.SubElement(annotation, 'folder').text = 'images'
    ET.SubElement(annotation, 'filename').text = image['file_name']
    ET.SubElement(annotation, 'path').text = os.path.abspath(image['file_name'])

    size = ET.SubElement(annotation, 'size')
    ET.SubElement(size, 'width').text = str(image['width'])
    ET.SubElement(size, 'height').text = str(image['height'])
    ET.SubElement(size, 'depth').text = '3'

    ET.SubElement(annotation, 'segmented').text = '0'

    for ann in annotations:
        obj = ET.SubElement(annotation, 'object')
        ET.SubElement(obj, 'name').text = categories_by_id[ann['category_id']]
        ET.SubElement(obj, 'pose').text = 'Unspecified'
        ET.SubElement(obj, 'truncated').text = '0'
        ET.SubElement(obj, 'difficult').text = '0'

        bbox = ann['bbox']
        bndbox = ET.SubElement(obj, 'bndbox')
        ET.SubElement(bndbox, 'xmin').text = str(int(bbox[0]))
        ET.SubElement(bndbox, 'ymin').text = str(int(bbox[1]))
        ET.SubElement(bndbox, 'xmax').text = str(int(bbox[0] + bbox[2]))
        ET.SubElement(bndbox, 'ymax').text = str(int(bbox[1] + bbox[3]))

    tree = ET.ElementTree(annotation)
    tree.write(annotation_filename)
```

```python
def r2(a, b, theta2):
    return a*b/np.sqrt((b*np.cos(theta2))**2 + (a*np.sin(theta2))**2)

def get_point(c,a,b,theta2):
    return (c[0]+r2(a,b,theta2)*np.cos(theta2), c[1]+r2(a,b,theta2)*np.sin(theta2))
```

```python
def get_bbox(indices, h,w):
    x1 = max(np.min(indices[:,0]), 0)
    y1 = max(np.min(indices[:,1]), 0)
    dy = np.max(indices[:,0])-x1
    dx = np.max(indices[:,1])-y1

    dy = min(dy, w-x1)
    dx = min(dx, h-y1)

    return x1, y1, dy, dx
```

```python
def get_img(img, img_path, id):
    h,w = img.size
    return {
            "date_captured": "",
            "neg_category_ids": [
            ],
            "id": id,
            "height": h,
            "width": w,
            "flickr_url": "",
            "coco_url": "",
            "not_exhaustive_category_ids": [
                1
            ],
            "file_name": img_path.split('/')[-1],
        }
    
```

```python
def get_annotation(x, img, id, image_id):
    sigma_x = (x.x2 - x.x1)//2
    sigma_y = (x.y2 - x.y1)//2

    indices = np.array([get_point((x.c_y, x.c_x), sigma_y+10, sigma_x+10, theta) for theta in np.linspace(0,2*np.pi,100)])

    bbox = list(get_bbox(indices, img.size[-1], img.size[-2]))

    indices = indices.reshape(1,-1).tolist()

    area = np.pi*(sigma_x+10)*(sigma_y+10)

    return {"area": area,
            "id": id,
            "image_id": image_id,
            "bbox": bbox,
            "category_id": 1,
            "segmentation": indices,
    }
```
